```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
```

```python
X = df.drop('Purchased', axis=1)
```

```python
y = df['Purchased']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_trainedscaled = scaler.transform(X_train)
```

```python
X_testscaled = scaler.transform(X_test)
```

```python
X_trainedscaled = pd.DataFrame(X_trainedscaled, columns=X_train.columns)
```

```python
X_testscaled = pd.DataFrame(X_testscaled, columns=X_train.columns)
```

```python
plt.scatter(X_train['Age'], X_train['EstimatedSalary'])
```

```python
plt.scatter(X_trainedscaled['Age'], X_trainedscaled['EstimatedSalary'], color = 'red')
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize = (8, 3))
sns.kdeplot(X_train['Age'], ax = ax1)
sns.kdeplot(X_train['EstimatedSalary'], ax = ax1)
sns.kdeplot(X_trainedscaled['Age'], ax=ax2)
sns.kdeplot(X_trainedscaled['EstimatedSalary'], ax = ax2)
plt.show()
```
