```python
# Ordinal Encoding
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('customer.csv', usecols=[2, 3, 4])
```

```python
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('purchased', axis=1), df['purchased'], test_size=0.2)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Poor', 'Average', 'Good'], ['School', 'UG', 'PG']])
```

```python
oe.fit(X_train)
```

```python
X_train.head(5)
```

```python
X_train_transformed = oe.transform(X_train)
X_test_transformed = oe.transform(X_test)
```

```python
X_train_transformed = pd.DataFrame(X_train_transformed, columns=X_train.columns)
X_test_transformed = pd.DataFrame(X_test_transformed, columns = X_test.columns)
```

```python
X_train_transformed.head(5)
```
