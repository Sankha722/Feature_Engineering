```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MaxAbsScaler
```

```python
df = pd.read_csv('wine_data.csv', usecols=[0,1,2])
```

```python
df.columns = ['level', 'Alcohol', 'Malic Acid']
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.describe()
```

```python
X = df.drop('level', axis=1)
y = df['level']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

```python
scaler = MaxAbsScaler()
```

```python
scaler.fit(X_train)
```

```python
X_trainscaled = scaler.transform(X_train)
X_testscaled = scaler. transform(X_test)
```

```python
X_trainscaled = pd.DataFrame(X_trainscaled, columns=X_train.columns)
X_testscaled = pd.DataFrame(X_testscaled, columns= X_test.columns)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(13, 5))
ax1.scatter(x=X_train['Alcohol'],y= X_train['Malic Acid'], c=y_train)
ax1.set_title('Before Max Absolute Scaling')
ax2.scatter(x=X_trainscaled['Alcohol'],y= X_trainscaled['Malic Acid'], c=y_train)
ax2.set_title('After Max Absolute Scaling')
```
